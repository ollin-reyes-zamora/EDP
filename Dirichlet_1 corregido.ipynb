**Ejercicio 1**

La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
Condiciones de frontera:
$$u(0,y)=u(1,y)=0, \;\;\; 0 < y < 1$$
$$u(x,0)=100, ~~ u(x,1)=200\;\;\;0<x < 1 $$
Serie de Fourier:
$$ u(x,y) = \sum_{n=1}^{\infty} sen(n \pi x) [ \frac{200 [ 1-(-1)^{n}]}{n \pi} \cosh (n \pi y) + \frac{200[1-(-1)^{n}][2- \cosh (n \pi)]}{(n \pi) \sinh (n\pi)} \sinh(n \pi y)] $$


```python
from pylab import *
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
def u_x_y(x,y):
    N = 10
    f = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          a = (200*(1-(-1)**(n)))/(n*pi)
          b = (200*(1-(-1)**(n))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += sin(n*pi*x)*(a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))
    return f

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="hot");
```
